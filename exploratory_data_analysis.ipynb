```python
sns.pairplot(merge[["Age", "Cites", "GSRank", "CitesPerYear", "CitesPerAuthor", "cited_by_patents_count", 
                    "cited_by_posts_count", "cited_by_accounts_count", "readers_count", "readers.mendeley", 
                    "cited_by_tweeters_count"]], diag_kind="kde", corner=True)
```

```python
merge['Cites_log'] = np.log10(merge['Cites']+1)
merge['CitesPerYear_log'] = np.log10(merge['CitesPerYear']+1)
merge['CitesPerAuthor_log'] = np.log10(merge['CitesPerAuthor']+1)
merge['cited_by_patents_count_log'] = np.log10(merge['cited_by_patents_count']+1)
merge['cited_by_posts_count_log'] = np.log10(merge['cited_by_posts_count']+1)
merge['cited_by_accounts_count_log'] = np.log10(merge['cited_by_accounts_count']+1)
merge['readers_count_log'] = np.log10(merge['readers_count']+1)
merge['readers.mendeley_log'] = np.log10(merge['readers.mendeley']+1)
merge['cited_by_tweeters_count_log'] = np.log10(merge['cited_by_tweeters_count']+1)
```

```python
sns.pairplot(merge[["Age", "Cites_log", "CitesPerYear_log", "CitesPerAuthor_log", "cited_by_patents_count_log", 
                    "cited_by_posts_count_log", "cited_by_accounts_count_log", "readers_count_log", "readers.mendeley_log", 
                    "cited_by_tweeters_count_log"]], diag_kind="kde", corner=True)
```
