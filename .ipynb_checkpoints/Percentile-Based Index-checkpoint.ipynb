```python
import pandas as pd
import seaborn as sns
import math
import numpy as np
```

```python
merge = pd.read_csv('/Users/maria/Desktop/ResearchImpactAI/cleaned_merged_dataset1.csv')
merge["Age"] = 2024-merge["Year"]
merge = merge.drop(index=[17992])
```

```python
merge.info(verbose=True)
```

# Bibliometrics section - Mendelay readers (25%), readers (25%), and cites (50%), all normalized for age

```python
merge["Cites_per_year"] = merge["Cites"]/merge["Age"]
merge["Cites_rank"] = merge.Cites_per_year.rank(pct = True)

merge["readers_count_per_year"] = merge["readers_count"]/merge["Age"]
merge["readers_count_rank"] = merge.Cites_per_year.rank(pct = True)

merge["readers.mendeley_per_year"] = merge["readers.mendeley"]/merge["Age"]
merge["readers.mendeley_rank"] = merge.Cites_per_year.rank(pct = True)
```

```python
merge["index_bibliometrics"] = 0.5*merge["readers.mendeley_rank"]+0.25*merge["readers_count_rank"]+0.25*merge["readers.mendeley_rank"]
```

```python
merge[merge["index_bibliometrics"]>0.95]
```

# Social media section - 33% cited_by_posts_count, 33% cited_by_accounts_count, 33% cited_by_tweeters_count  

```python
merge["tweets_per_year"] = merge["cited_by_tweeters_count"]/merge["Age"]
merge["tweets_rank"] = merge.tweets_per_year.rank(pct = True)

merge["posts_per_year"] = merge["cited_by_posts_count"]/merge["Age"]
merge["posts_rank"] = merge.posts_per_year.rank(pct = True)

merge["accounts_per_year"] = merge["cited_by_accounts_count"]/merge["Age"]
merge["accounts_rank"] = merge.accounts_per_year.rank(pct = True)
```

```python
merge["index_social_media"] = (1/3)*merge["tweets_rank"]+(1/3)*merge["posts_rank"]+(1/3)*merge["accounts_rank"]
```

```python
merge[merge["index_social_media"]>0.95]
```

# News media  

```python
merge["msm_per_year"] = merge["cited_by_msm_count"]/merge["Age"]
merge["news_index"] = merge.msm_per_year.rank(pct = True)
```

# Patents

```python
merge["patent_per_year"] = merge["cited_by_patents_count"]/merge["Age"]
merge["patent_index"] = merge.patent_per_year.rank(pct = True)
```

# Policies

```python
merge["policies_per_year"] = merge["cited_by_policies_count"]/merge["Age"]
merge["policies_index"] = merge.policies_per_year.rank(pct = True)
```

# Overall - 24% bibliometrics, 15% social media, 15% news media, 23% patents, 23% policies, 10% journal impact factor

```python
merge["overall_index"] = 0.24*merge["index_bibliometrics"]+0.15*merge["index_social_media"]+0.15*merge['news_index']+0.23*merge['patent_index']+0.23*merge['policies_index']
```

```python
merge[merge["overall_index"]>0.98]
```

```python
merge.to_csv("merged_data_with_indices.csv")
```
