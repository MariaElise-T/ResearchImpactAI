```python
import pandas as pd
import seaborn as sns
import math
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
merge = pd.read_csv('THRC_final_indices.csv')
```

```python
plt = sns.barplot(data=merge.sort_values(by="Short_Name"), x="Short_Name", y="overall_index")
plt.set_xticklabels(plt.get_xticklabels(), rotation=45)
plt.axhline(0.5, ls='--', c='red')
plt.set(xlabel ="", ylabel = "Overall Index", title ='THRC FF Papers')
```
