```python
import pandas as pd
import numpy as np
from impact_factor.core import Factor
```

```python
crossref = pd.read_csv('TXST_crossref/TXST_crossref_2000.csv')  
years = np.arange(2001, 2025, 1)
for year in years:
    filepath = 'TXST_crossref/TXST_crossref_'+str(year)+'.csv'
    papers = pd.read_csv(filepath)  
    crossref = pd.concat([crossref, papers])
papers = pd.read_csv('TXST_crossref/THRC papers.csv')  
crossref = pd.concat([crossref, papers])
```

```python
crossref = crossref.fillna(0)
```

```python
crossref.shape
```

```python
crossref = crossref.reset_index()
crossref = crossref.drop(['index', 'CitesURL', 'QueryDate', 'GSRank', 'CitationURL'], axis=1)
crossref["impact_factor"] = None
```

```python
crossref.shape
```

```python
crossref
```

```python
factor = Factor()
for i in range(0, len(crossref)):
    if i % 100 == 0:
        print("On iteration ", i)
    if crossref.loc[i, 'Source'] != 0:
        factor_dict = factor.search(crossref.loc[i, 'Source'])
        if len(factor_dict)>0:
            crossref.loc[i, 'impact_factor'] = factor_dict[0]['factor']
        else:
            crossref.loc[i, 'impact_factor'] = 0
    else:
        crossref.loc[i, 'impact_factor'] = 0
```

```python
crossref.shape
```

```python
crossref.to_csv("crossref_allyears.csv", index=False)
crossref.shape
```
