```python
import pandas as pd
import numpy as np
import requests
import json
import datetime
import ast
```

```python
inputs = pd.read_csv('TXST_crossref/TXST_crossref_2000.csv')  
years = np.arange(2001, 2025, 1)
for year in years:
    filepath = 'TXST_crossref/TXST_crossref_'+str(year)+'.csv'
    papers = pd.read_csv(filepath)  
    inputs = pd.concat([inputs, papers])
papers = pd.read_csv('TXST_crossref/THRC papers.csv')  
inputs = pd.concat([inputs, papers])
```

```python
len(inputs)
```

```python
hits=0
total=0
json_altmetrics = list()
for doi in inputs['DOI']:
    total+=1
    altmetric_url = 'https://api.altmetric.com/v1/doi/'+str(doi)
    response = requests.get(altmetric_url)
    if(response.status_code==200):
        json_altmetrics.append(response.content)
        hits+=1
    if(total%100==0):
        print(total)
        print(datetime.datetime.now())
print("Total entries found: ", hits)
print("Hit rate is :", hits/total)
```

```python
len(json_altmetrics)
```

```python
with open('altmetric_data.json', 'w') as f:
    for item in json_altmetrics:
        json.dump(item.decode('utf-8'), f)
```
