```python
import pandas as pd
import numpy as np
from scipy.stats import rankdata
import matplotlib.pyplot as plt

# === STEP 1: Load your data ===
df = pd.read_csv("allsources_percentiles.csv")  # Replace with your actual filename

# === STEP 2: Define impact areas and weights ===
impact_areas = {
    "Bibliometrics": [
        "citation.Scopus_plumx", "citation.CrossRef_plumx", "citation.Academic Citation Index (ACI) - airiti_plumx",
        "citation.PubMed_plumx", "capture.READER_COUNT_plumx", "mention.ABSTRACT_VIEWS_plumx",
        "mention.DOWNLOAD_COUNT_plumx", "mention.FULL_TEXT_VIEWS_plumx", "capture.EXPORTS_SAVES_plumx",
        "readers.citeulike_altmetrics", "readers.mendeley_altmetrics", "readers.connotea_altmetrics",
        "readers_count_altmetrics"
    ],
    "Social Media": [
        "cited_by_posts_count_altmetrics", "cited_by_accounts_count_altmetrics",
        "cited_by_tweeters_count_altmetrics", "mention.FACEBOOK_COUNT_plumx",
        "cited_by_fbwalls_count_altmetrics", "cited_by_linkedin_count_altmetrics",
        "cited_by_weibo_count_altmetrics"
    ],
    "News Media": [
        "cited_by_msm_count_altmetrics", "mention.NEWS_COUNT_plumx"
    ],
    "Patents": [
        "cited_by_patents_count_altmetrics", "citation.Patent Families_plumx"
    ],
    "Policies": [
        "citation.Policy Citation_plumx", "cited_by_policies_count_altmetrics"
    ],
    "Journal Impact Factor": ["impact_factor"],
    "Misc.": [
        "cited_by_feeds_count_altmetrics", "cited_by_videos_count_altmetrics",
        "cited_by_wikipedia_count_altmetrics"
    ]
}

weights = {
    "Bibliometrics": 0.50,
    "Social Media": 0.10,
    "News Media": 0.10,
    "Patents": 0.10,
    "Policies": 0.10,
    "Journal Impact Factor": 0.05,
    "Misc.": 0.05
}

# === STEP 3: Compute subindex averages ===
for area, variables in impact_areas.items():
    df[f"{area}_avg"] = df[variables].mean(axis=1)

# === STEP 4: Convert subindices to percentile ranks ===
for area in impact_areas:
    df[f"{area}_pct"] = rankdata(df[f"{area}_avg"], method='average') / len(df)

# === STEP 5: Weighted composite score ===
df["composite_score"] = sum(
    df[f"{area}_pct"] * weight for area, weight in weights.items()
)

# === STEP 6: Final percentile rank ===
df["composite_index_pct"] = rankdata(df["composite_score"], method='average') / len(df)
```

```python
df.to_csv("df_percentiles.csv", index=False)
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 10))
axes = axes.flatten()

for i, area in enumerate(impact_areas):
    axes[i].hist(df[f"{area}_pct"], bins=10, edgecolor='black', alpha=0.8)
    axes[i].set_title(area)
    axes[i].set_xlabel("Percentile Rank")
    axes[i].set_ylabel("Frequency")

# Hide any unused subplot slots
for j in range(len(impact_areas), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.suptitle("Subindex Percentile Distributions", fontsize=16, y=1.02)
plt.savefig("subindex_percentiles.png", dpi=300, bbox_inches="tight")
plt.close()

# === PLOT 2: Overall Composite Index ===
plt.figure(figsize=(8, 6))
plt.hist(df["composite_index_pct"], bins=10, edgecolor='black', alpha=0.9)
plt.title("Overall Composite Index Percentile Distribution")
plt.xlabel("Percentile Rank")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("composite_index_percentile.png", dpi=300)
plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# === Load and merge data (adjust filenames if needed) ===
cluster_df = pd.read_csv("df_with_cluster_info.csv", index_col=0)
pareto_df = pd.read_csv("df_pareto.csv", index_col=0)
composite_df = pd.read_csv("df_percentiles.csv", index_col=0)

# Merge into a single DataFrame
merged = cluster_df.copy()
merged["pareto_rank"] = pareto_df["pareto_rank"]
merged["composite_index_pct"] = composite_df["composite_index_pct"]
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# === Define sets using merged DataFrame ===

# Set 1: Top 500 by composite index percentile (higher is better)
top_500_composite = set(
    merged.sort_values("composite_index_pct", ascending=False).head(500).index
)

# Set 2: All items where cluster rank == 0 (best cluster-based rank)
rank0_cluster = set(merged[merged["rank"] == 0].index)

# Set 3: All items where pareto_rank == 0 (on Pareto front)
pareto_front0 = set(merged[merged["pareto_rank"] == 0].index)

# === Plot Venn Diagram ===
plt.figure(figsize=(8, 6))
venn3(
    [top_500_composite, rank0_cluster, pareto_front0],
    set_labels=["Top 500 Composite", "Cluster Rank = 0", "Pareto Front = 0"]
)
plt.title("Overlap of Top 500 Composite, Cluster Rank 0, and Pareto Front 0")
plt.tight_layout()
plt.savefig("venn_top500_cluster0_pareto0.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# === Calculate mean composite index percentile by cluster rank ===
cluster_means = (
    merged.groupby("rank")["composite_index_pct"]
    .mean()
    .sort_index()
)

# === Plot the means ===
plt.figure(figsize=(10, 6))
cluster_means.plot(kind="bar", edgecolor="black")
plt.xlabel("Cluster Rank")
plt.ylabel("Mean Composite Index Percentile")
plt.title("Mean Composite Index Percentile by Cluster Rank")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig("mean_percentile_by_cluster_rank.png", dpi=300)
plt.show()
```
